```python
import numpy as np
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import io
import requests
import seaborn as sns

%matplotlib inline
import ssl
# следующая строчка подключает сертификат для защищенного соединения
ssl._create_default_https_context = ssl._create_unverified_context
df=pd.read_csv('1.csv')
df.head()
```

Пример 1

```python
fonds = df['fonds'] # показатель позитивных эмоций
product = df['product']
#plt.scatter(freedom, positive)
plt.scatter(fonds, product)
plt.xlabel('Стоимость фондов')
plt.ylabel('Выработка на одного сотрудника')
```

```python
fonds = np.array(fonds).reshape(-1,1)
product = np.array(product).reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    fonds,
    product,
    test_size = 0.2
)
model.fit(X_train, y_train)
predicted_y = model.predict(X_test)
plt.xlabel('Direct_Bilirubin')
plt.ylabel('Total_Bilirubin')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-50, 200, 100)
# Посчитаем предсказанные значения
# y = kx + b
# k = model.coef_
# b = model.intercept_
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(-10,80)
plt.ylim(-10,80)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
print('k=',model.coef_)
print('b=',model.intercept_)
```
